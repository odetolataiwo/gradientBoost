Your challenge is to scrape data from this link: https://en.wikipedia.org/wiki/List_of_African_countries_by_area


Focus on the first table:
Take rank, country, and area.

Using pandas split country area into two columns: km**2 and sq mi.

Load the columns into a DataFrame.

```python
# importing libraries
from bs4 import BeautifulSoup
import requests
```

```python
link = 'https://en.wikipedia.org/wiki/List_of_African_countries_by_area'

# use request to get the raw HTML content from the website
HTML_script = requests.get(link)

# We now extract the text from the HTML_script
text_script = HTML_script.text

useful_text = BeautifulSoup(text_script, "html.parser") 
#We can now print the useful_text using prettify module.
print(useful_text.prettify())
```

```python
table = useful_text.find('table')
table_rows = table.find_all('tr')
table_rows
```

```python
table_rows[1].find('td').text.rstrip()
```

```python
table_rows[1].find('a')
```

```python
rank = [row.find('td').text for row in table_rows[1:-1]]
print(rank)
```

```python
rank = [r.rstrip() for r in rank]
print(rank)
```

```python
countries = [row.find('a').text for row in table_rows[1:-1]]
print(countries)
```

```python
area = [row.find_all('td')[2].text for row in table_rows[1:-1]]
print(area)
```

```python
area = [x.replace('\xa0km2 (',' ').replace('\xa0sq\xa0mi)\n','').split(' ')  for x in area]
print(area)
```

```python
import pandas as pd
df = pd.DataFrame()
```

```python
df['Rank'] = rank
df['Country'] = countries
df['Area (km**2)'] = [i[0] for i in area]
df['Area (sq mi)'] = [i[1] for i in area]

df.index = df['Rank']
df.drop('Rank',inplace=True, axis=1)

df
```
